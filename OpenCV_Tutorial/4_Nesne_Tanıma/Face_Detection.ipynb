```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# haar cascades kullanarak yüz tespiti yapmak için gerekli kütüphaneleri import ediyoruz
```

```python
erkek = cv2.imread(r"opencv_2\media\man.jpg",0)
plt.imshow(erkek,cmap="gray")
```

```python
kadin = cv2.imread(r"opencv_2\media\woman.jpg",0)
plt.imshow(kadin,cmap="gray")
```

```python
toplu = cv2.imread(r"opencv_2\media\bilim_adamlari.jpg",0)
plt.imshow(toplu,cmap="gray")   
```

```python
face_cascade = cv2.CascadeClassifier(r"opencv_2\haar_cased\frontalface.xml")
# yüz tespiti yapmak için kullanılan xml dosyasını import ediyoruz
```

```python
def detect_face(img): # yüz tespiti yapmak için fonksiyon oluşturuyoruz
    face_img = img.copy()
    
    face_recs = face_cascade.detectMultiScale(face_img)  # yüz tespiti yapılıyor
    
    for (x,y,w,h) in face_recs: # yüzün etrafına dikdörtgen çiziliyor
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
        
    return face_img
```

```python
result = detect_face(toplu)
```

```python
plt.imshow(result,cmap="gray")
```

```python
result = detect_face(erkek)
```

```python
plt.imshow(result,cmap="gray")
```

```python
result = detect_face(kadin)
plt.imshow(result,cmap="gray")
```
