```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flatt_chess = cv2.imread(r'opencv_2\media\chess.jpg')
flatt_chess = cv2.cvtColor(flatt_chess,cv2.COLOR_BGR2RGB)
plt.imshow(flatt_chess)
```

```python
found, corners = cv2.findChessboardCorners(flatt_chess,(7,7))
# findChessboardCorners = köşelri bulur
```

```python
corners.shape
```

```python
flatt_chess_copy = flatt_chess.copy()
cv2.drawChessboardCorners(flatt_chess_copy,(7,7),corners,found) # köşeleri çizer
```

```python
plt.imshow(flatt_chess_copy)
```

```python
dots = cv2.imread(r'opencv_2\media\dot.jpg')
```

```python
plt.imshow(dots)
```

```python
found, corners = cv2.findCirclesGrid(dots,(10,10),cv2.CALIB_CB_SYMMETRIC_GRID)
```

```python
db_image_circles = dots.copy()
cv2.drawChessboardCorners(db_image_circles,(10,10),corners,found)
```
