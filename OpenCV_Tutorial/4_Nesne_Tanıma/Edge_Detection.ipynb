```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread(r"opencv_2\media\dog_head2.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img, threshold1=127, threshold2=127)
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img, threshold1=0, threshold2=127)
plt.imshow(edges)
```

```python
blur_img = cv2.blur(img, ksize=(5,5))
edges = cv2.Canny(image=blur_img, threshold1=0, threshold2=90)
plt.imshow(edges)
```
