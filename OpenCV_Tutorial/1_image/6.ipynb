```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread(r"C:\Users\Yasin\Desktop\opencv_2\bird.jpg")
plt.imshow(img)
```

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
cv2.rectangle(img2, (1200,650), (2450,2000), (0,255, 0), 35)
plt.imshow(img2)
```

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

noktalar= np.array([[1200,650],[1200,2000],[2400,1300]], np.int32)
cv2.fillPoly(img2, [noktalar],(255, 0, 0))
plt.imshow(img2)
```
