```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
empty_tables = np.zeros((512,512, 3), dtype=np.int16)
plt.imshow(empty_tables)
```

```python
cv2.putText(empty_tables,text="MERHABA", org=(10,450), fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScale=2, color=(0,0,255), thickness=6)
plt.imshow(empty_tables)
```

```python
noktalar = np.array([[100, 300],[200,200], [400, 300], [200, 400]], dtype=np.int32)
cv2.polylines(empty_tables, pts=[noktalar],isClosed=True, color=(255,0,255), thickness=5)
plt.imshow(empty_tables)
```
