```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
empty_tables = np.zeros(shape=(512, 512, 3), dtype=np.int16)
plt.imshow(empty_tables)
```

```python
cv2.rectangle(empty_tables, (200, 200), (300,300), (200, 250, 5),thickness=3)
plt.imshow(empty_tables)
```

```python
cv2.circle(empty_tables, center=(100, 100), radius=50, color=(100, 200,100), thickness=-1)
plt.imshow(empty_tables)
```

```python
cv2.ellipse(empty_tables, center=(400, 100),axes=(50,30),angle=0,startAngle=0,endAngle=360, color=(255, 10, 255), thickness=-1)
plt.imshow(empty_tables)
```

```python
cv2.line(empty_tables, pt1=(0,0), pt2=(512, 512), color=(0, 0, 255), thickness=5)
plt.imshow(empty_tables)
```
