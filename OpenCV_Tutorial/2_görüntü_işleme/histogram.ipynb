```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bird = cv2.imread(r"opencv_2\media\bird.jpg")
show_bird = cv2.cvtColor(bird, cv2.COLOR_BGR2RGB)

wall = cv2.imread(r"opencv_2\media\wall4.jpg")
show_wall = cv2.cvtColor(wall, cv2.COLOR_BGR2RGB)

rainbow = cv2.imread(r"opencv_2\media\gokkusagi.jpg")
show_rainbow = cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(show_bird)
```

```python
plt.imshow(wall)
```

```python
plt.imshow(show_rainbow)
```

```python
hist_values =cv2.calcHist([wall], channels=[0], mask=None, histSize=[256], ranges=[0,256])
```

```python
plt.plot(hist_values)
```

```python
hist_values =cv2.calcHist([bird], channels=[0], mask=None, histSize=[256], ranges=[0,256])
plt.plot(hist_values)
```

```python
img = show_bird
```

```python
color = ('b', 'g', 'r')

for i, col in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(histr, color=col)
    plt.xlim([0, 256])

plt.title("HISTOGRAM")
    
```
