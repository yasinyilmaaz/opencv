```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12, 10))
    ax =fig.add_subplot(111)
    ax.imshow(img, cmap="gray")
```

```python
image = cv2.imread(r"opencv_2\media\sudoku.jpg",0)
display_img(image)
```

```python
sobenX = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5) # dikey izgileri daha belli eder
# cv2.CV_64F = yapmak olduumuz ilemin kalite deeri
display_img(sobenX)
```

```python
sobelY = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5) # yatay izgileri daha belli eder
display_img(sobelY)
```

```python
laplacian = cv2.Laplacian(image, cv2.CV_64F)
display_img(laplacian)
```

```python
blended = cv2.addWeighted(sobenX, 0.5, sobelY, 0.5, 0)
display_img(blended)
```

```python
ret, threshold = cv2.threshold(laplacian, 25, 255, cv2.THRESH_BINARY)
display_img(threshold)
```

```python
kernel = np.ones((2, 2), np.uint8)
gradient = cv2.morphologyEx(laplacian, cv2.MORPH_OPEN, kernel)
display_img(gradient)
```

```python
ret, threshold = cv2.threshold(gradient, 120, 255, cv2.THRESH_BINARY)
display_img(threshold)
```
