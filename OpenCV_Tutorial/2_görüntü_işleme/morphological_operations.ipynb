```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# Gürültüyü azaltmak için kullanılır
```

```python
# görüntüyü oluşturmak için
def load_img():
    blank_img = np.zeros((600, 600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img, text="ABCDE", org=(50, 300), fontFace=font, fontScale=5, color=(255, 255, 255), thickness=25)
    return blank_img   
```

```python
#  görüntüyü göstermek için
def display_img(img):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap="gray")
```

```python
i = load_img()
display_img(i)
```

```python
kernel = np.ones((5, 5), dtype=np.uint8) # 5x5 lik bir kernel oluşturduk
result = cv2.erode(i, kernel, iterations=1)
# iterasyon = 1 ise 1 kez erode işlemi yapar
display_img(result)
```

```python
img = load_img()
white_noise = np.random.randint(0, 2, size=(600, 600)) * 255 # gürültü oluşturduk
noise_img = white_noise + img
display_img(noise_img)
```

```python
opening = cv2.morphologyEx(noise_img, cv2.MORPH_OPEN, kernel) # gürültüyü azalttık
# cv2.MORPH_OPEN = erosion + dilation
display_img(opening)
```

```python
img = load_img()
black_noise = np.random.randint(0,2, size=(600,600)) * -255 # gürültü oluşturduk
black_noise_img = img + black_noise #her piksele gürültü ekledik
```

```python
black_noise_img[black_noise_img == -255] = 0 # -255 olanları 0 yaptık
```

```python
black_noise_img.min()
```

```python
display_img(black_noise_img)
```

```python
clossing = cv2.morphologyEx(black_noise_img, cv2.MORPH_CLOSE, kernel)
display_img(clossing)
```

```python
img = load_img()
display_img(img)
```

```python
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
display_img(gradient)
```
