```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap=cmap)    
```

```python
img = cv2.imread(r"opencv_2\media\fil.jpg")
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(img_gray,127, 255, cv2.THRESH_BINARY) 
display_img(thresh,cmap='gray')
```

```python
img = cv2.imread(r"opencv_2\media\fil.jpg")
img_HSV =cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
display_img(img_gray)
```

```python
kernel = np.ones((4, 4), dtype=np.float32) /10
kernel
```

```python
d2 = cv2.filter2D(img_gray, -1, kernel)
display_img(d2)
```

```python
sobel = cv2.Sobel(img_gray, cv2.CV_64F, 1, 0, ksize=5)
display_img(sobel)
```

```python
kernel = np.ones((5, 5), dtype=np.uint8)
gradient = cv2.morphologyEx(img_gray, cv2.MORPH_GRADIENT, kernel)
display_img(gradient)
```
