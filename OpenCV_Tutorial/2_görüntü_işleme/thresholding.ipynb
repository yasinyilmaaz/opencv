```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread(r"opencv_2\media\gokkusagi.jpg",0)
plt.imshow(img, cmap="gray")
```

```python
img.max()
```

```python
img.min()
```

```python
ret, threshold =cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
# ret = threshold degeri
# cv2.THRESH_BINARY = eşik değerinin üstündeki değerleri 255 yapar, altındaki değerleri 0 yapar
# cv2.THRESH_BINARY_INV = eşik değerinin üstündeki değerleri 0 yapar, altındaki değerleri 255 yapar
plt.imshow(threshold, cmap="gray")
```

```python
img = cv2.imread(r"opencv_2\media\bulmaca.jpg",0)
plt.imshow(img, cmap="gray")
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap="gray")
```

```python
show_pic(img)
```

```python
ret , th1= cv2.threshold(img, 127, 255, cv2.THRESH_BINARY) # 127 eşik değeri
plt.imshow(th1, cmap="gray")  
```

```python
img_2 = cv2.imread(r"opencv_2\media\bulmaca2.jpg",0)
plt.imshow(img_2, cmap="gray")
```

```python
ret, th3 = cv2.threshold(img_2, 127, 255, cv2.THRESH_BINARY)
plt.imshow(th3, cmap="gray")
```

```python
th4 = cv2.adaptiveThreshold(img_2,255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 8)
# adaptiveThreshold = adaptif eşik değeri
# cv2.ADAPTIVE_THRESH_MEAN_C = ortalama değerini alır
# 11 = komşuluk değeri
# 8 = sabit değer
plt.imshow(th4, cmap="gray")
```

```python
th5 = cv2.adaptiveThreshold(img_2,255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11, 8) # gauss
plt.imshow(th5, cmap="gray")
```
