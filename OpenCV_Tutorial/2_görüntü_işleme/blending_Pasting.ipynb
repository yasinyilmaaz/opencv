```python
img = cv2.imread(r"opencv_2\media\bird.jpg")
img_warning = cv2.imread(r"opencv_2\media\warning3.jpg")
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_warning_RGB = cv2.cvtColor(img_warning, cv2.COLOR_BGR2RGB)
img_warning_RGB = cv2.resize(img_warning_RGB, (600, 600))
```

```python
plt.imshow(img_RGB)
```

```python
plt.imshow(img_warning_RGB)
```

```python
large_img = img_RGB
small_img = img_warning_RGB

large_img[0:600, 0:600] = small_img
plt.imshow(large_img)
```

```python
img_RGB_roi = img_RGB[600:2000, 1100:2500] 
plt.imshow(img_RGB_roi)
```

```python
gray = cv2.cvtColor(img_warning, cv2.COLOR_RGB2GRAY)
gray = cv2.resize(gray, (600, 600))
plt.imshow(gray, cmap="gray")
```

```python
plt.imshow(cv2.bitwise_not(gray), cmap="gray")
```

```python
import numpy as np
msk_inv = cv2.bitwise_not(gray)
```

```python
white_background = np.full((600,600,3), 255, dtype=np.uint8)
white_background.shape
```

```python
bk = cv2.bitwise_or(white_background,white_background, mask=msk_inv)
bk.shape
```

```python
plt.imshow(bk, cmap="gray")
```

```python
fg = cv2.bitwise_or(img_warning_RGB, img_warning_RGB, mask=msk_inv)
```

```python
plt.imshow(fg)
```

```python
fg_r = cv2.resize(fg, (1400, 1400))
final_roi = cv2.bitwise_or(img_RGB_roi,fg_r)
plt.imshow(final_roi)
```

```python
img_RGB[600:2000, 1100:2500] = final_roi
plt.imshow(img_RGB)
```
