```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

```

```python
def load_img():
    img = cv2.imread(r"opencv_2\media\wall2.jpg")
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
i = load_img()
display_img(i)
```

```python
gamma = 0.9
```

```python
result = np.power(i, gamma) 
plt.imshow(result)
# display_img(result)
```

```python
img = load_img()
cv2.putText(img, text="duvar", org=(10,100), fontFace=cv2.FONT_HERSHEY_COMPLEX, fontScale=2, color=(255,0,0), thickness=2, lineType=cv2.LINE_AA)
display_img(img)
```

```python
blur = cv2.blur(img, ksize=(5, 5))
display_img(blur)
```

```python
bird = cv2.imread(r"opencv_2\media\bird.jpg",0)
bird = cv2.resize(bird, (600, 600))
white_noise = np.random.randint(0, 2, size=(600, 600)) * 255
bird_noise = bird + white_noise
display_img(bird_noise)
```

```python
blur = cv2.blur(bird_noise, ksize=(11,11))
display_img(blur)
```
