```python
import keyboard # klavyeden giriş almak için
import uuid # ekrandan alınan resimleri isimlendirmek için
import time  # zaman işlemleri için
from PIL import Image 
from mss import mss #  
```

```python
# http://www.trex-game.skipser.com/

mon = {"top": 518, "left": 670, "width": 250, "height": 100} # oyun alanının alınacağı alanın koordinatları
sct = mss() # istenen alandaki görüntüyü alır

i = 0 # alınan görüntülerin isimlendirilmesi için
def record_screen(record_id, key): # alınan görüntüyü kaydeder
    global i
    i += 1
    print(f"{key}: {i}")
    img = sct.grab(mon) # oyun alanındaki görüntüyü alır
    im = Image.frombytes("RGB", img.size, img.rgb) # alınan görüntüyü resme çevirir
    im.save(f"./img/{record_id}_{key}_{i}.png") # resmi kaydeder

is_exit = False # çıkış yapılıp yapılmadığını kontrol etmek için

def exit(): # çıkış yapmak için
    global is_exit
    is_exit = True
keyboard.add_hotkey("esc", exit) # esc tuşuna basıldığında çıkış yapar

record_id = uuid.uuid4() # alınan görüntülerin isimlendirilmesi için id oluşturur
while True:
    if is_exit: # çıkış yapılıp yapılmadığını kontrol eder
        break
    try:
        if keyboard.is_pressed(keyboard.KEY_UP):
            record_screen(record_id, "up") # yukarı tuşuna basıldığında görüntüyü kaydeder
        elif keyboard.is_pressed(keyboard.KEY_DOWN):
            record_screen(record_id, "down") # aşağı tuşuna basıldığında görüntüyü kaydeder
        elif keyboard.is_pressed("right"): # iki durum da gerçekleşmediğinde herhangi bir hata almamak için eklendi
            record_screen(record_id, "right") # sağ tuşuna basıldığında görüntüyü kaydeder
        time.sleep(0.1)
    except RuntimeError:
        continue
```
