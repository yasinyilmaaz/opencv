```python
from keras import models
import numpy as np
from PIL import Image
import keyboard
import time
from mss import mss
import os

mon = {"top": 528, "left": 670, "width": 250, "height": 100}
sct = mss()

widht = 125
height = 100

# model Yükleme
model = models.model_from_json(open(os.path.join(os.getcwd(), "model.json"), "r").read())
model.load_weights(os.path.join(os.getcwd(), "trex_weight.h5"))

# down => 0, right => 1, up=> 2
labes = ["Down", "Right", "Up"]

framerate_time = time.time() # zaman tutmak için
countour = 0 # 
i = 0 #
delay = 0.4 # bekleme
key_down_pressed = False # aşağı basılıp basılmadığını kontrol eder
while True:

    img = sct.grab(mon) # ekran görüntüsü alır
    im = Image.frombytes("RGB", img.size, img.rgb) 
    im2 = np.array(im.convert("L").resize((widht, height)))
    im2 = im2 / 255

    X = np.array([im2])
    X =X.reshape(X.shape[0], widht, height, 1)
    r = model.predict(X) # eğitilmiş model ile tahmini

    result = np.argmax(r) # en yüksek değeri tahminin indexini alır

    if result == 0: # down => 0
        keyboard.press(keyboard.KEY_DOWN)
        key_down_pressed = True # basılı tutulduğunu belirtir

    elif result == 2: # up => 2
        if key_down_pressed:
            keyboard.release(keyboard.KEY_DOWN)
            key_down_pressed = False
        time.sleep(delay)
        keyboard.press(keyboard.KEY_UP)
        
        # zıpladıktan sonra belli frameler arasında belli bekletme süreleri belirledik
        if i < 1500:
            time.sleep(0.3)
        elif 1500 < i and i < 5000:
            time.sleep(0.2)
        else:
            time.sleep(0.17)
        
        keyboard.press(keyboard.KEY_DOWN)
        keyboard.release(keyboard.KEY_DOWN)
    countour += 1
    if (time.time() - framerate_time) > 1:
        countour = 0
        framerate_time = time.time()
        if i <= 1500:
            delay*= 0.003
        else:
            delay -= 0.005
        if delay < 0:
            delay = 0
        
        print("----------------------")
        print("Down: {}\n Right: {}\n Up: {}\n".format(r[0][0], r[0][1], r[0][2]))



        
```
