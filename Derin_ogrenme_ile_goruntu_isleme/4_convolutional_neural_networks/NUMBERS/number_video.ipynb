```python
import cv2
import pickle # model yüklemek için
import numpy as np
from keras import models
import os
```

```python
def preProcess(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = cv2.resize(img, (32, 32))
    img = img / 255.0
    img = img.reshape(1, 32, 32, 1)
    return img

cap = cv2.VideoCapture(0)
cap.set(3, 480)
cap.set(4, 480)


# pickle_in = open(r"C:\Users\90505\Desktop\model_egitimi\convolutional_neural_networks\NUMBERS\model_trained_new.p","rb")
# model = pickle.load(pickle_in)

# model = models.model_from_json(open(r"C:\Users\90505\Desktop\model_egitimi\convolutional_neural_networks\NUMBERS\model_trained.json", "r").read())
# model.load_weights(r"C:\Users\90505\Desktop\model_egitimi\convolutional_neural_networks\NUMBERS\model_trained.h5")
import tensorflow as tf

# # JSON formatındaki modeli yükleme
# with open(r"C:\Users\90505\Desktop\model_egitimi\convolutional_neural_networks\NUMBERS\model_trained.json", "r") as json_file:
#     model_json = json_file.read()

# model = tf.keras.models.model_from_json(model_json)

with open(os.path.join(os.getcwd(), "model_trained.p"), "rb") as file:
    model = pickle.load(file)


while True:
    success, frame = cap.read()

    img = np.asarray(frame)
    img = cv2.resize(img, (32, 32))
    img = preProcess(img)

    img = img.reshape(1,32,32,1)

    # Predict
    # classIndex = int(model.predict_classes(img))
    predictions = model.predict(img)
    classIndex = int(np.argmax(predictions))
    propVal = np.amax(predictions)
    # print(classIndex, propVal)

    print(propVal)
    if propVal > 0.5:
        cv2.putText(frame, "Predicted Number: " + str(classIndex), (20, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 1)
    
    cv2.imshow("Rakam Siniflandirma", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

```
